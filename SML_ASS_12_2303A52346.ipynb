```python
import pandas as pd
df=pd.read_csv('/content/breast_cancer_survival.csv')
df
```

```python
df['Gender']=df['Gender'].replace(['MALE','FEMALE'],[0,1])
df['Tumour_Stage']=df['Tumour_Stage'].replace(['I','II','III'],[1,2,3])
df['ER status']=df['ER status'].replace(['Positive','Negative'],[1,0])
df['PR status']=df['PR status'].replace(['Positive','Negative'],[1,0])
df['HER2 status']=df['HER2 status'].replace(['Positive','Negative'],[1,0])
df['Patient_Status']=df['Patient_Status'].replace(['Alive','Dead'],[1,0])
df['Histology']=df['Histology'].replace(['Infiltrating Ductal Carcinoma','Infiltrating Lobular Carcinoma','Mucinous Carcinoma'],[1,0,2])
df['Surgery_type']=df['Surgery_type'].replace(['Lumpectomy','Modified Radical Mastectomy','Simple Mastectomy','Other'],[0,1,2,3])
df
```

```python
df.fillna(0,inplace=True)
df
```

```python
target=df['Patient_Status']
target
```

```python
features=df.drop(['Patient_Status','Date_of_Surgery','Date_of_Last_Visit'],axis=1)
features
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(features,target,test_size=0.2,random_state=1)
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier()
model.fit(x_train,y_train)
y_pre=model.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pre)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.utils import resample
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

logistic_model = LogisticRegression()
svc_model = SVC()
knn_model = KNeighborsClassifier()

logistic_scores = []
svc_scores = []
knn_scores = []

for i in range(0,1000):
    x_bootstrap, y_bootstrap = resample(x_train, y_train)
    logistic_model.fit(x_bootstrap, y_bootstrap)
    y_pred = logistic_model.predict(x_test)
    logistic_scores.append(accuracy_score(y_test, y_pred))

for i in range(0,1000):
    x_bootstrap, y_bootstrap = resample(x_train, y_train)
    svc_model.fit(x_bootstrap, y_bootstrap)
    y_pred = svc_model.predict(x_test)
    svc_scores.append(accuracy_score(y_test, y_pred))

for i in range(0,1000):
    x_bootstrap, y_bootstrap = resample(x_train, y_train)
    knn_model.fit(x_bootstrap, y_bootstrap)
    y_pred = knn_model.predict(x_test)
    knn_scores.append(accuracy_score(y_test, y_pred))

plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.hist(logistic_scores, bins=20)
plt.title('Logistic Regression')
plt.xlabel('Accuracy')
plt.ylabel('Frequency')

plt.subplot(1, 3, 2)
plt.hist(svc_scores, bins=20)
plt.title('Support Vector Machine')
plt.xlabel('Accuracy')

plt.subplot(1, 3, 3)
plt.hist(knn_scores, bins=20)
plt.title('K-Nearest Neighbors')
plt.xlabel('Accuracy')

plt.tight_layout()
plt.show()
```
