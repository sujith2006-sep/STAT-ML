```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("/content/sample_data/california_housing_test.csv")

columns_to_analyze = ['latitude', 'total_rooms', 'housing_median_age', 'total_bedrooms']
bins_list = [5, 10, 15, 20]

for col in columns_to_analyze:
    for bins in bins_list:
        plt.hist(data[col], bins=bins, density=True, alpha=0.6, label=f'Bins: {bins}')
        plt.xlabel(col)
        plt.ylabel('Density')
        plt.title(f'Density Estimation for {col}')
        plt.legend()
        plt.show()


        print(f"Observations for {col} with {bins} bins:")
        print("-" * 30)
```
