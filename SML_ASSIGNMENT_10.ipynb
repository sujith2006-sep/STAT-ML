```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
import random
```

# Question 1:

```python
df = pd.read_csv('/content/breast_cancer_survival.csv')
df
```

```python
df['Gender'] = df['Gender'].replace({'MALE': 0, 'FEMALE': 1})
df['ER status'] = df['ER status'].replace({'Positive': 1, 'Negative': 0})
df['PR status'] = df['PR status'].replace({'Positive': 1, 'Negative': 0})
df['HER2 status'] = df['HER2 status'].replace({'Positive': 1, 'Negative': 0})
df['Tumour_Stage'] = df['Tumour_Stage'].replace({'I': 1, 'II': 2, 'III': 3, 'IV': 4})
df['Patient_Status'] = df['Patient_Status'].replace({'Alive': 1, 'Dead': 0})
df.fillna(0,inplace = True)
```

```python
data = df.drop('Date_of_Surgery',axis = 1)
data = data.drop('Date_of_Last_Visit',axis = 1)
data = data.drop('Histology',axis = 1)
data = data.drop('Surgery_type',axis = 1)
```

```python
x = data.drop('Patient_Status',axis = 1)
y = data['Patient_Status']
```

```python
count = 0
count_ = 0
for i in df['Patient_Status']:
  if(i == 1.0):
    count = count + 1
  elif(i == 0.0):
    count_ = count_ + 1

print(count)
print(count_)
y.isnull().sum()
```

```python
i = 0
while(i<9):
  j = random.uniform(0.5,0.9)
  x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = j)
  model = KNeighborsClassifier()
  model.fit(x_train,y_train)
  yp = model.predict(x_test)
  accu = accuracy_score(y_test,yp)
  print("Iteration No:",i+1)
  print(accu)
  print(classification_report(y_test,yp))
  print(confusion_matrix(y_test,yp))
  i = i + 1
```

# Question 2:

```python
from sklearn.svm import SVC
lt = [0.2,0.25,0.3,0.35]
for i in lt:
  x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = i)
  model = SVC()
  model.fit(x_train,y_train)
  yp = model.predict(x_test)
  accu = accuracy_score(y_test,yp)
  print("Iteration No:",lt.index(i)+1)
  print(accu)
  print(classification_report(y_test,yp))
  print(confusion_matrix(y_test,yp))
```
